```python
# Q1)Write a Python function to sum all the numbers in a list

def sum_num(num_of_lst):
    sum_of_total = 0
    for x in num_of_lst:
        sum_of_total += x
    return sum_of_total
print(sum_num((8,2,3,0,7)))
```

```python
# Q2)Write a Python program to reverse a string.
#Sample String : "1234abcd" Expected Output : "dcba4321"

def reverse_strg(string):
    if len(string) == 0:
        return string
    else:
        return reverse_strg(string[1:]) + string[0]

reverse_strg("1234abcd")
```

```python
#Q 3) Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.
#Sample String : 'The quick Brow Fox' Expected Output : No. of Upper case characters : 3 No. of Lower case Characters : 12

def string(y):
    t={"uppercase":0, "lowercase":0}
    for k in y:
        if k.isupper():
            t["uppercase"]+=1
        elif k.islower():
            t["lowercase"]+=1
        else:
               pass
    
    print ("total Upper case  :", t["uppercase"])
    print ("total Lower case  :", t["lowercase"])

string('The quick Brow Fox')
```
