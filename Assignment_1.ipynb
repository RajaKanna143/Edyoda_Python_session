### Q1) Write a Python program to get the Fibonacci series between 0 to 50

```python
a_num,b_num=0,1
while b_num<=50:
    print(b_num)
    a_num,b_num=b_num,a_num+b_num
```

### Q2) Write a Python program that accepts a word from the user and reverse it.

```python
str_ch = input('Please enter a word :')

if str_ch:
    print('Reverse of the word is :', str_ch[::-1])
```

### Q3) Write a Python program to count the number of even and odd numbers from a series of numbers.

```python
numb = (1, 2, 3, 4, 5, 6, 7, 8, 9)
even_numb = 0
odd_numb = 0
for num in numb:
    if num % 2 == 0:
        even_numb += 1
    else:
        odd_numb += 1
        
print('Total even numbers are :', even_numb)
print('Total odd numbers are :', odd_numb )
```
